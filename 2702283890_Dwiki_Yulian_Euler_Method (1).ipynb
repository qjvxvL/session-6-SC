```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Define parameters
f = lambda t, s: np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step size
h = 0.01
# Numerical grid
t = np.arange(0, 1 + h, h)
# Initial condition
y0 = 1

# Define the function for the differential equation y' = x^2 - 4y
def f(x, y):
    return x**2 - 4*y

# Explicit Euler Method
y = np.zeros(len(t))
y[0] = y0

for i in range(0, len(t) - 1):
    y[i + 1] = y[i] + h * f(t[i], y[i])

# Plot the approximate solution
plt.figure(figsize=(12, 8))
plt.plot(t, y, 'b--', label='Approximate')

# Note: We plot the exact solution for reference.
# To plot the exact solution, you'd need to solve the ODE analytically,
# but here we focus on the numerical solution using Euler's method.

plt.title('Approximate Solution for ODE $y\' + 4y = x^2$ using Euler\'s Method')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step size
h = 0.1
# Numerical grid
t = np.arange(0, 0.5 + h, h)  # Adjusted the end point to 0.5
# Initial condition
y0 = 1

# Define the function for the differential equation y' = sin(y)
def f(x, y):
    return np.sin(y)

# Explicit Euler Method
y = np.zeros(len(t))
y[0] = y0

for i in range(0, len(t) - 1):
    y[i + 1] = y[i] + h * f(t[i], y[i])

# Plot the approximate solution
plt.figure(figsize=(12, 8))
plt.plot(t, y, 'b--', label='Approximate')

plt.title('Approximate Solution for ODE $y\' = \sin(y)$ using Euler\'s Method')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()

print(f"Approximate solution at x = 0.5: {y[-1]:.4f}")
```
